```python
import requests
import pandas as pd

food_prices = []
url = "https://microdata.worldbank.org/index.php/api/tables/data/fcv/wld_2021_rtfp_v02_m?limit=1000&offset=0&country=Kenya&fields=o_maize,h_maize,l_maize,c_maize,inflation_maize,trust_maize,o_sorghum,h_sorghum,l_sorghum,c_sorghum,trust_sorghum,o_potatoes,h_potatoes,l_potatoes,trust_potatoes,c_potatoes,o_food_price_index,h_food_price_index,l_food_price_index,c_food_price_index,lat,lon,geo_id,start_dense_data,last_survey_point,year,month,mkt_name,DATES,currency"
response = requests.get(url)
if response.status_code == 200:
    prices = response.json()
    for i in prices["data"]:
        food_prices.append(i)
    print("Data extraction complete!")
    print(food_prices[0])
else:
    print(f"Error extracting data from API: {response.status_code}, {response.text}")
```

```python
# convert data into dataframe
df = pd.DataFrame(food_prices)
df.head(15)
```

```python
df.drop(["trust_sorghum", "trust_potatoes", "trust_maize", "last_survey_point", "start_dense_data"], axis=1, inplace=True)
df
```

```python
df.rename(columns={"DATES": "date"}, inplace=True)
df
```

```python
df["date"] = pd.to_datetime(df["date"])
df.dtypes
```

```python
df.drop("inflation_maize", axis=1, inplace=True)
df.iloc[0]
```

```python
# load into test database table
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine

load_dotenv()
#print(os.getenv("PSQL_URI"))
engine = create_engine(os.getenv("PSQL_URI"))
try:
    df.to_sql('test_food_prices', con=engine, index=False, schema='foodprices', if_exists='append')
    print("Data loaded into PostgreSQL successfully")
except Exception as e:
    print(f"Error loading into database: {e}")
```
